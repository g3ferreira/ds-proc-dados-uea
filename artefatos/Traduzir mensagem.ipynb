## Traduzindo os comentários

```
translator = Translator()
```

Função para traduzir cada linha do dataframe

```
def traduzir(comentario) :
    try :
        comentario_traduzido = translator.translate(comentario, dest='en').text
    except:
        comentario_traduzido = numpy.nan
    finally:
        print(comentario_traduzido)
        time.sleep(2)
        return comentario_traduzido
```

Loop para salvar as traduções do campo reviewBody no arquivo a cada 20 traduções feitas

```
i = 90000
j = 90020
while(i < j and j <= 100000) :
  print("----------", i, "a", j,"----------")
  df_reviewBodyEn = df_dataset.loc[i:j-1, ('reviewBody')].apply(traduzir)
  i = j
  j = j + 20
  df_reviewBodyEn.to_csv('/content/drive/MyDrive/UEA - Pós Ciência de Dados/2. Programação para Ciência de Dados/ds-proc-dados-uea/artefatos/dados/reviewBodyEn - 90.000 a 99.999.csv', mode='a', header=False)
  print("Salvo")
```
