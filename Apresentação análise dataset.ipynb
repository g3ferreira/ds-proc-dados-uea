**Resumo**

Este Google Colab tem por finalidade apresentar um resumo do
dataset e suas estatísticas, mostrando quais dados estão presentes nele, através do uso das bibliotecas pandas e matplotlib.

**Importando a biblioteca Pandas e Matplotlib**

```
import pandas as pd
import matplotlib.pyplot as plt
```

**Lendo o Dataset**

```
df_dataset = pd.read_json("http://tiagodemelo.info/datasets/dataset-v2.dat", lines=True)
```

**Imprimindo o dataset**

```
df_dataset
```

**Language**

```
df_language = df_dataset.language.value_counts(ascending=False)
```

```
df_language
```

```
df_language.plot(kind='bar')
```

**ratingValue**

```
df_ratingValue = df_dataset.ratingValue.value_counts(ascending=False)
```

```
df_ratingValue
```

```
df_ratingValue.plot(kind='bar')
```

**itemReviewed**

```
df_normalize = pd.json_normalize(df_dataset.itemReviewed)
```

```
df_normalize
```

```
df_normalize.columns = ['name', 'website', 'email', 'phoneNumber', 'averageRating', 'streetAddress', 'addressLocality', 'addressRegion', 'postalCode', 'country']
```

```
df_normalize
```

```
df_addressRegion = df_normalize.addressRegion.value_counts(ascending=False)
```

```
df_addressRegion
```

```
df_addressRegion.plot(kind='bar')
```

```
df_name = df_normalize.name.value_counts(ascending=False)
```

```
df_name
```

```
df_averageRating = df_normalize.averageRating.value_counts(ascending=False)
```

```
df_averageRating
```

```
df_averageRating.plot(kind='bar')
```
